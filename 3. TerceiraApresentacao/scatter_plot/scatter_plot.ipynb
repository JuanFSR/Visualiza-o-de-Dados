```python
import pandas as pd
import numpy as np
df = pd.read_csv('../data/pre_processado_data.csv', sep=',')
df.head()
```

```python
def find_area(empresa):
    areas = df.query(f'`Nome Fantasia` == "{empresa}"')['Área']
    areas_unique = areas.unique()
    areas_count = [[area, np.count_nonzero(np.where(areas == area))] for area in areas_unique]
    maior = [areas_unique[0], 0]
    for area in areas_count:
        if area[1] > maior[1]:
            maior[0] = area[0]
            maior[1] = area[1]
    return maior[0]
```

```python
empresas = df['Nome Fantasia'].unique()
data = [[find_area(empresa), empresa, np.mean(df.query(f'`Nome Fantasia` == "{empresa}" & `Nota do Consumidor` != -1.0')['Tempo Resposta']), np.mean(df.query(f'`Nome Fantasia` == "{empresa}" & `Nota do Consumidor` != -1.0')['Nota do Consumidor']), df.query(f'`Nome Fantasia` == "{empresa}" & `Nota do Consumidor` != -1.0').count()[0]] for empresa in empresas]
data
```

```python
processed_df = pd.DataFrame(data, columns=['Área', 'Empresa', 'Tempo Resposta', 'Nota do Consumidor', 'Reclamações'])
processed_df.head()
```

```python
import plotly.express as px

fig = px.scatter(processed_df, x="Nota do Consumidor", y="Tempo Resposta",
                 size="Reclamações", color="Área",
                 hover_name="Empresa", log_x=True, size_max=100)
fig.show()
```

```python
fig.write_html("scatter_plot.html")
```
