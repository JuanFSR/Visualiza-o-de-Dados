```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/finalizadas_2021-11.csv', sep=';')
# df.columns
df.tail()
```

# Pré-processamento
Abaixo será realizado a identificação de atributos faltantes, realizando o tratamento que será descrito logo em seguida. Como não há uma grande dimensionalidade no nosso dataset, não foi necessário realizar uma diminuição da dimensionalidade.

```python
dados_faltantes = df.isnull().sum()
print(dados_faltantes)
```

```python
# Verificando quantidade de dados faltantes na coluna e retirando-os caso necessário
limite = df.shape[0]*0.85
drop_colunas = []
quantidade_null = df.isnull().sum()
colunas = df.columns

for quantidade, col in zip(quantidade_null, colunas):
    if quantidade >= limite:
        drop_colunas.append(col)

print(drop_colunas)
```

```python
limite_linha = df.shape[1]*0.5
df.dropna(thresh=limite_linha, axis=0, inplace=True)

shape_inicial = df.shape
print(shape_inicial)
```

# Mapeamento de dados nominais para números

Será utilizado o mapeamento de dimensões nominais para números paras colunas: Situação, Avaliação Reclamação.

```python
situacoes = df['Situação'].unique()
avaliacao_reclamacao = df['Avaliação Reclamação'].unique()

print("* A coluna Situação contém os seguintes valores: ")
[print("\t -" + sit) for sit in situacoes]

print("* A coluna Avaliação Reclamação contém os seguintes valores ", avaliacao_reclamacao)
for ar in avaliacao_reclamacao:
    print("\t -" + ar)
```

```python
# Realizando mapeamento
print(shape_inicial)

df['Finalizada avaliada'] = df['Situação'].replace('Finalizada avaliada', 1)
df['Finalizada avaliada'] = df['Finalizada avaliada'].replace('Finalizada não avaliada', 0)

df['Finalizada não avaliada'] = df['Situação'].replace("Finalizada não avaliada", 1)
df['Finalizada não avaliada'] = df['Finalizada não avaliada'].replace("Finalizada avaliada", 0)

df.drop('Situação', axis=1)
```

```python
# Realizando mapeamento da coluna Avaliação Reclamação
df['Resolvida'] = df['Avaliação Reclamação'].replace('Resolvida', 1)
df['Resolvida'] = df['Resolvida'].replace('Não Avaliada', 0)
df['Resolvida'] = df['Resolvida'].replace('Não Resolvida', 0)

df['Não Avaliada'] = df['Avaliação Reclamação'].replace('Não Avaliada', 1)
df['Não Avaliada'] = df['Não Avaliada'].replace('Resolvida', 0)
df['Não Avaliada'] = df['Não Avaliada'].replace('Não Resolvida', 0)

df['Não Resolvida'] = df['Avaliação Reclamação'].replace('Não Resolvida', 1)
df['Não Resolvida'] = df['Não Avaliada'].replace('Resolvida', 0)
df['Não Resolvida'] = df['Não Avaliada'].replace('Não Avaliada', 0)

df.drop('Avaliação Reclamação', axis=1)

print("A estrutura antes de realizar o mapeamento era: ", shape_inicial)
print("A estrutura depois de realizar o mapeamento é: ", df.shape)
```

# Tratamento dos valores faltantes

Para valores:
* Numéricos: É possível substituir assinalar um valor sentinela, calcular um substituto através da média ou assinalando um valor baseado no vizinho mais próximo.

* Categóricos: É possível substituir pelo valor de maior ocorrência na coluna, ou simplemente remover a linha que contém este atributo faltante.
